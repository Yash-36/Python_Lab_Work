### 08) Define Time class with hour and minute as data member. Also define addition method to add two time objects.

```python
class Time:
    
    def __init__(self,h,m):
        self.hour = h
        self.minute = m
        
    def addTime(self,t1,t2):
        temp_hour = t1.hour + t2.hour
        temp_minute = t1.minute + t2.minute
        if(temp_minute >= 60):
            temp_hour = temp_hour + 1
            temp_minute = temp_minute - 60
        self.hour=temp_hour
        self.minute=temp_minute

obj1 = Time(2,10)
obj2 = Time(3,20)

obj3 = Time(0,0)
obj3.addTime(obj1,obj2)

print(f"HH : MM  =  {obj3.hour} : {obj3.minute}")
```

### 10) Create a child class Bus that will inherit all of the variables and methods of the Vehicle class

    def __init__(self, name, max_speed, mileage):
        self.name = name
        self.max_speed = max_speed
        self.mileage = mileage
        
Create a Bus object that will inherit all of the variables and methods of the parent Vehicle class and display it.        

```python
class Vehicle:
    def __init__(self, name, max_speed, mileage):
        self.name = name
        self.max_speed = max_speed
        self.mileage = mileage
class Bus(Vehicle):
    def printDetails(self):
        print(f"Name : {super().name}")
        print(f"Max Speed : {super().max_speed}")
        print(f"Mileage : {super().mileage}")
        
b1 = Bus("ferrari",200,50)
b1.printDetails()
    
```
